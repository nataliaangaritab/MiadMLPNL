## Datos predicción probabilidad de admisión a la universidad

En este taller se usará el conjunto de datos de admisiones a una universidad proveniente de la base de datos de Kaggle. Cada observación contiene la probabilidad de que un estudiante sea admitido por la universidad, dadas distintas variables predictoras como el puntaje del examén TOEFL y GRE, el promedio (GPA), entre otras. El objetivo es predecir la probabilidad de admissión de cada estudiante. Para más detalles pueden visitar el siguiente enlace: [datos](https://www.kaggle.com/mohansacharya/graduate-admissions).

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importación librerías
import numpy as np
import keras 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from keras import initializers
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Carga de datos de archivo .csv
data = pd.read_csv('https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/datasets/universityGraduateAdmissions.csv', index_col=0)
data.head()
```

```python
# Hitogramas de las varaibles del dataset
data.hist(bins=25, figsize=(15,15))
plt.show()
```

```python
# Definición de variables predictoras (X)
X = data.drop(data.columns[-1], axis=1)
# Definición de variable de interés (y)
Y = data[data.columns[-1]]
```

```python
Y = pd.DataFrame(Y)
```

```python
# Relación entre predictores y variable objetivo
fig, ax = plt.subplots(2,3,figsize=(15, 8))
j=0
for i in [0,1,5]:
    ax[0,j].scatter(X.iloc[:,i],Y)
    j=j+1
j=0
for i in [2,3,4]:
    ax[1,j].scatter(X.iloc[:,i],Y)
    j=j+1
```

```python
# Separación de variables predictoras (X) y variable de interés (y) en set de entrenamiento y test usandola función train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3, random_state=22)
```

```python
# Normalización de variables predictoras (X) con la función StandardScaler
from sklearn.preprocessing import StandardScaler

# Definición de la función StandardScaler
scaler = StandardScaler()

# Transformación de los set de entrenamiento y test
X_train = pd.DataFrame(data=scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test = pd.DataFrame(data=scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

X_train = np.array(X_train)
Y_train = np.array(Y_train)

X_test = np.array(X_test)
Y_test = np.array(Y_test)
```

```python
# Definición de dimensiones de salida, variables de interés
output_var = Y_train.shape[1]
print(output_var, ' output variables')
```

```python
# Definición de dimensiones de entrada, varaibles predictoras
dims = X_train.shape[1]
print(dims, 'input variables')
```

## Punto 1 - Red Neuronal de una capa

En la celda 1 creen una **red neuronal de una capa** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes, usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador y el número de épocas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Importación librerías
from keras.layers import Activation
from livelossplot import PlotLossesKeras
from keras import backend as K
import keras.optimizers as opts
```

```python
# Celda 1

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()
# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(output_var, input_shape=(dims,)))

# Impresión de la arquitectura de la red neuronal
print(model.summary())
```

```python
# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
# Entrenamiento de la red neuronal con 50 épocas
model.fit(X_train, Y_train, 
          verbose=1, 
          epochs=300, 
          validation_data=(X_test, Y_test),
          callbacks=[PlotLossesKeras()])
```

```python
# Evaluación del modelo
from sklearn.metrics import mean_squared_error, mean_absolute_error

pred1 = model.predict(X_test)

MAE1 = (mean_absolute_error(Y_test,pred1))
MSE1 = (mean_squared_error(Y_test,pred1))
print("Mean Absolute Error (MAE): {:.4f}".format(mean_absolute_error(Y_test, pred1)))
print("Mean Squared Error (MSE): {:.4f}".format(mean_squared_error(Y_test, pred1)))
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test, pred1)
plt.plot(Y_test, Y_test)
plt.show()
```

Con el fin definir, entrenar y evaluar el modelo de regresión utilizando una red neuronal: 

- Se utiliza el modelo secuencial (Sequential()) de Keras para definir la arquitectura de la red neuronal. 
- Se agrega una  capa densa (Dense()) con un tamaño de salida (output_var) y una forma de entrada (input_shape) especificada por las dimensiones (dims). Se imprime la  arquitectura de la red neuronal con el metodo summary() del modelo, mostrando detalle de la capa.
- Se compila el modelo utilizando el optimizador Adam, ya que según la literatura es el que ofrece mejores resultados. También se uso el método de ensayo y error por las épocas. La función de pérdida 'mean_squared_error' debe usarse para este caso donde el problema es una regresion, teniendo en cuanto que queremos predecir una probabilidad (valor continuo).
- Se entrena el modelo utilizando el método fit(). Se especifica el conjunto de entrenamiento (X_train, Y_train) 70%, y validación (X_test, Y_test) 30%.
- Se grafica la pérdida del modelo durante el entrenamiento utilizando la biblioteca Matplotlib. La gráfica muestra la pérdida en función de las épocas para los conjuntos de entrenamiento y validación, lo que permite evaluar la convergencia y el rendimiento del modelo. En este caso, se trabaja con 300 épocas para visualizar más datos pero se puede observar que los resultados convergen después de 100 épocas.
- Se utiliza el conjunto de prueba para realizar predicciones (model.predict(X_test)) y se calculan las métricas de error, incluyendo el error absoluto medio (MAE) 0.0495 y el error cuadrático medio (MSE) 0.0040.

**OPTIMIZER: ADAM**

## Punto 2 - Red Neuronal multicapa

En la celda 2 creen una **red neuronal con dos capas** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador, el número de épocas y el número de neuronas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 2

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(256, input_shape=(dims,),activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(output_var))
model.add(Activation('relu'))

# Impresión de la arquitectura de la red neuronal
print(model.summary())
```

```python
# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
# Separación de datos de entrenamiento para considerar un set de validación durante entrenamiento
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.15, random_state=42)
```

```python
# Entrenamiento de la red neuronal con 300 épocas
model.fit(X_train, Y_train,
          verbose=1, 
          epochs=300, 
          validation_data = (X_val, Y_val),
          callbacks=[PlotLossesKeras()])
```

```python
pred2 = model.predict(X_test)

MAE2 = (mean_absolute_error(Y_test,pred2))
MSE2 = (mean_squared_error(Y_test,pred2))
print("Mean Absolute Error (MAE): {:.4f}".format(mean_absolute_error(Y_test, pred2)))
print("Mean Squared Error (MSE): {:.4f}".format(mean_squared_error(Y_test, pred2)))
```

```python
plt.scatter(Y_test, pred2)
plt.plot(Y_test, Y_test)
plt.show()
```

Con el fin definir, entrenar y evaluar el modelo de regresión utilizando dos capas con la librería Keras:
- Se utiliza el modelo secuencial (Sequential()) de Keras para definir la arquitectura de la red neuronal. 
- Se agregan dos capas densas (Dense()). La primera capa tiene 256 neuronas de salida, utiliza una función de activación relu y toma como entrada un tensor de forma input_shape=(dims,).
- La segunda capa tiene 256 neuronas de salida y utiliza también una función de activación relu. Se imprime la  arquitectura de la red neuronal con el metodo summary() del modelo, mostrando detalle de la capa.
- Se eligen 256 neuronas, ya que se observo que una mayor cantidad de neuronas provocaba overfiting en el modelo mientras que menos neuronas hacia que el modelo tuviese dificultades para aprender pues el error era mayor (se hacen pruebas con 50, 100, 150, 200, 300).
- Se compila el modelo utilizando el optimizador Adam, siguiendo las sugerencias en la literatura como se explico en el punto anterior. Se empleo la función de pérdida 'mean_squared_error' debido a que nuestro problema es de regresion.
- Se entrena el modelo utilizando el método fit(). Se especifica el conjunto de entrenamiento (X_train, Y_train) 70%,  y validación (X_test, Y_test) 30%.
- Se realiza un nuevo split con data de validación, para que el algoritmo logre generalizar todos los casos posibles.
- Se grafica la pérdida del modelo durante el entrenamiento utilizando la biblioteca Matplotlib. La gráfica muestra la pérdida en función de las épocas para los conjuntos de entrenamiento y validación, lo que permite evaluar la convergencia y el rendimiento del modelo. Nuevamente, se eligen 300 épocas para visualizar diferentes escenarios y en este caso hay variabilidad pues no convergen datos de training y validation.
- Se utiliza el conjunto de prueba para realizar predicciones (model.predict(X_test)) y se calculan las métricas de error, incluyendo el error absoluto medio (MAE) 0.0565 y el error cuadrático medio (MSE) 0.0062.

## Punto 3  - Red Neuronal multicapa

En la celda 3 creen **una red neuronal con más de una capa con la librería Keras, usando early stopping y dropout,** que prediga la probabilidad de admisión de los estudiantes con los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador, el número de épocas y el número de neuronas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 3

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(256, input_shape=(dims,),activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(output_var))
model.add(Activation('relu'))

# Impresión de la arquitectura de la red neuronal
print(model.summary())
```

```python
# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
# Separación de datos de entrenamiento para considerar un set de validación durante entrenamiento
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.15, random_state=42)
```

```python
# Definición de la función EarlyStopping para considerar durante el entrenamiento
from keras.callbacks import EarlyStopping, ModelCheckpoint

early_stop = EarlyStopping(monitor='val_loss', patience=50, verbose=1)
```

```python
# Definición de la función ModelCheckpoint para guardar el modelo con mejor desempeño
fBestModel = 'best_model.h5'
best_model = ModelCheckpoint(fBestModel, verbose=0, save_best_only=True)
```

```python
# Entrenamiento de la red neuronal con 300 épocasy early stopping
model.fit(X_train, Y_train, 
          verbose=True, 
          epochs=300, 
          batch_size=128,
          validation_data=(X_val, Y_val),
          callbacks=[best_model, early_stop, PlotLossesKeras()])
```

```python
pred3 = model.predict(X_test)

MAE3 = (mean_absolute_error(Y_test,pred3))
MSE3 = (mean_squared_error(Y_test,pred3))
print("Mean Absolute Error (MAE): {:.4f}".format(mean_absolute_error(Y_test, pred3)))
print("Mean Squared Error (MSE): {:.4f}".format(mean_squared_error(Y_test, pred3)))
```

```python
plt.scatter(Y_test, pred3)
plt.plot(Y_test, Y_test)
plt.show()
```

Con el fin definir, entrenar y evaluar el modelo de regresión utilizando una red neuronal: 

- Se utiliza el modelo secuencial (Sequential()) de Keras para definir la arquitectura de la red neuronal. 
- La primera capa densa tiene 256 neuronas y utiliza la función de activación relu. El parámetro input_dim se establece según el número de columnas en X_train.
- Se añade una capa de Dropout con una tasa de 0.25 para regularizar la red neuronal y evitar el sobreajuste.
- Luego se agrega una capa densa con 256 neuronas de salida y la función de activación relu.
- Se añade otra capa de Dropout con una tasa de 0.25.
- Finalmente, se agrega una capa densa con un número de neuronas de salida igual a output_var, que no está especificado en el código proporcionado.
- En esta red neural se incremento el numero de neuronas en cada capa ya que con el dropout el 25% de ellas estaran aleatoriamente dormidas durante el entrenamiento. El numero de neuronas propuestas se eligio por medio de ensayo y error, buscando que el modelo aprendiera sin presentar sobreentrenamiento pues se relizan pruebas con diferentes valores y el error fue mayor (50, 100, 150, 200, 300).
- Se imprime la  arquitectura de la red neuronal con el metodo summary() del modelo, mostrando detalle de la capa. 
- Se compila el modelo utilizando el optimizador Adam, siguiendo lo sugerido en la literatura. La función de pérdida 'mean_squared_error' ya que el problema es de regresion.
- Se utiliza el callback EarlyStopping para detener el entrenamiento si no se produce una mejora en la función de pérdida durante un número determinado de épocas (patience=60 en este caso). Esto ayuda a evitar el sobreajuste y mejora la eficiencia del entrenamiento.
- Se entrena el modelo utilizando el método fit(). Se especifica el conjunto de entrenamiento (X_train, Y_train) 70%,  y validación (X_test, Y_test) 30%.
- Se utiliza el conjunto de prueba para realizar predicciones (model.predict(X_test)) y se calculan las métricas de error, incluyendo el error absoluto medio (MAE) 0.0843 y el error cuadrático medio (MSE) 0.0092.
- Se grafica la pérdida del modelo durante el entrenamiento utilizando la biblioteca Matplotlib. La gráfica muestra la pérdida en función de las épocas para los conjuntos de entrenamiento y validación, lo que permite evaluar la convergencia y el rendimiento del modelo.

## Punto 4 - Comparación y análisis de resultados

En la celda 4 comparen los resultados obtenidos de las diferentes redes y comenten las ventajas del mejor modelo y las desventajas del modelo con el menor desempeño.

```python
# Celda 4

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Comparación de modelos MSE ")
ejeX = ['Red Neuronal una capa', 'Red Neuronal multi', 'Red Neuronal multi early dropout']
ejeY = [round(MSE1,5), round(MSE2,5), round(MSE3,5)]
ax.bar(ejeX,ejeY)
def addlabels(x,y,plotP):
    for i in range(len(x)):
        plotP.text(i,y[i],y[i])
addlabels(ejeX,ejeY,plt)
plt.show()
```

La red neuronal con una sola capa presenta un error cuadrático medio (MSE) menor a los demás.

En este caso, las diferencias en el MSE entre las diferentes redes son pequeñas porque se evaluan valores muy pequeños como la probabilidad. Esto sugiere que agregar capas adicionales a la red neuronal no proporciona mejoras significativas en la precisión de las predicciones. El rendimiento de la red neuronal con varias capas es comparable al de la red neuronal de una sola capa. Esto sugiere que la complejidad adicional de las capas adicionales no aporta beneficios notables en la precisión del modelo. 

Un estudio hecho al principio de este cuaderno muestra la relación ente las variables predictoras y la variable objetivo. En los graficos se puede observar como la relación para todas las variables es más u menos lineal, esto explica el hecho de que aumentar la complejidad de la red neural no aumente significativamente su desempeño.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Comparación de modelos MAE ")
ejeX = ['Red Neuronal una capa', 'Red Neuronal multi', 'Red Neuronal multi early dropout']
ejeY = [round(MAE1,5), round(MAE2,5), round(MAE3,5)]
ax.bar(ejeX,ejeY)
def addlabels(x,y,plotP):
    for i in range(len(x)):
        plotP.text(i,y[i],y[i])
addlabels(ejeX,ejeY,plt)
plt.show()
```

La red neuronal con una sola capa presenta un error cuadrático absoluto (MAE) menor pero muy cercano al modelo multicapa. De acuerdo a los resultados, el modelo no mejora cuuando se hace más complejo como se explica anteriormente.

En este caso, las diferencias en el MSE y MAE para los modelos son pequeñas. Esto sugiere que agregar capas adicionales a la red neuronal no proporciona mejoras significativas en la precisión de las predicciones. El rendimiento de la red neuronal con varias capas es comparable al de la red neuronal de una sola capa en términos de precisión. Esto sugiere que la complejidad adicional de las capas adicionales no aporta beneficios notables en la precisión del modelo.

También se puede observar que el modelo de multicapa el menos estable, ya que no hay una clara convergencia entre el training y el validation. Para terminar, es importante mencionar que el menor validation loss (0.0040) lo tiene el modelo de una capa, esto quiere decir que la discrepancia entre las salidas predichas por el modelo y las salidas reales presentan menor variación.
