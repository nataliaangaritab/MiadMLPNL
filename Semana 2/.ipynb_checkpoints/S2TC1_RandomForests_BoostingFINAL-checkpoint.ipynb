### Punto 1 - Árbol de decisión manual

En la celda 1 creen un árbol de decisión **manualmente**  que considere los set de entrenamiento y test definidos anteriormente y presenten el RMSE y MAE del modelo en el set de test.

```python
# Celda 1
# Definición de la función que calcula el gini index
def gini(y):
    if y.shape[0] == 0:
        return 0
    else:
        return 1 - (y.mean()**2 + (1 - y.mean())**2)

# Definición de la función gini_imputiry para calular la ganancia de una variable predictora j dado el punto de corte k
def gini_impurity(X_col, y, split):
    
    filter_l = X_col < split
    y_l = y.loc[filter_l]
    y_r = y.loc[~filter_l]
    
    n_l = y_l.shape[0]
    n_r = y_r.shape[0]
    
    gini_y = gini(y)
    gini_l = gini(y_l)
    gini_r = gini(y_r)
    
    gini_impurity_ = gini_y - (n_l / (n_l + n_r) * gini_l + n_r / (n_l + n_r) * gini_r)
    
    return gini_impurity_
```

```python
# Definición de la función best_split para calcular cuál es la mejor variable y punto de cortepara hacer
# la bifurcación del árbol
def best_split(X, y, num_pct=10):
    
    features = range(X.shape[1])
    
    best_split = [0, 0, 0]  # j, split, gain
    
    # Para todas las varibles 
    for j in features:
        
        splits = np.percentile(X.iloc[:, j], np.arange(0, 100, 100.0 / (num_pct+1)).tolist())
        splits = np.unique(splits)[1:]
        
        # Para cada partición
        for split in splits:
            gain = gini_impurity(X.iloc[:, j], y, split)
                        
            if gain > best_split[2]:
                best_split = [j, split, gain]
    
    return best_split
```

```python
# Definición de la función tree_grow para hacer un crecimiento recursivo del árbol
def tree_grow(X, y, level=0, min_gain=0.001, max_depth=None, num_pct=10):
    
    # Si solo es una observación
    if X.shape[0] == 1:
        tree = dict(y_pred=y.iloc[:1].values[0], level=level, split=-1, n_samples=1, gain=0)
        return tree
    
    # Calcular la mejor división
    j, split, gain = best_split(X, y, num_pct)
    
    # Guardar el árbol y estimar la predicción
    y_pred = int(y.mean())
    
    tree = dict(y_pred=y_pred, level=level, split=-1, n_samples=X.shape[0], gain=gain)
    # Revisar el criterio de parada 
    if gain < min_gain:
        return tree
    if max_depth is not None:
        if level >= max_depth:
            return tree   
    
    # Continuar creando la partición
    filter_l = X.iloc[:, j] < split
    X_l, y_l = X.loc[filter_l], y.loc[filter_l]
    X_r, y_r = X.loc[~filter_l], y.loc[~filter_l]
    tree['split'] = [j, split]

    # Siguiente iteración para cada partición
    
    tree['sl'] = tree_grow(X_l, y_l, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    tree['sr'] = tree_grow(X_r, y_r, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    
    return tree
```

```python
decisionTreeManual = tree_grow(X_train, y_train, level=0, min_gain=0.001, max_depth=3, num_pct=10)
decisionTreeManual
```

```python
# Definición de la función tree_predict para hacer predicciones según las variables 'X' y el árbol 'tree'

def tree_predict(X, tree, proba=False):
    
    predicted = np.ones(X.shape[0])

    # Revisar si es el nodo final
    if tree['split'] == -1:
        if not proba:
            predicted = predicted * tree['y_pred']
        else:
            predicted = predicted * tree['y_prob']
            
    else:
        j, split = tree['split']
        filter_l = (X.iloc[:, j] < split)
        X_l = X.loc[filter_l]
        X_r = X.loc[~filter_l]

        if X_l.shape[0] == 0:  # Si el nodo izquierdo está vacio solo continua con el derecho 
            predicted[~filter_l] = tree_predict(X_r, tree['sr'])
        elif X_r.shape[0] == 0:  #  Si el nodo derecho está vacio solo continua con el izquierdo
            predicted[filter_l] = tree_predict(X_l, tree['sl'])
        else:
            predicted[filter_l] = tree_predict(X_l, tree['sl'])
            predicted[~filter_l] = tree_predict(X_r, tree['sr'])

    return predicted
```

```python
y_pred = tree_predict(X_test, decisionTreeManual)

# RMSE
print("RMSE: %.2f" % mean_squared_error(y_test, y_pred, squared=False))
# MAE
print("MAE: %.2f" % mean_absolute_error(y_test, y_pred))
```

### Punto 2 - Bagging manual

En la celda 2 creen un modelo bagging **manualmente** con 10 árboles de regresión y comenten sobre el desempeño del modelo.

```python
# Crear una submuestra aleatoria del conjunto de datos con reemplazo
def subsample(x,y, ratio):
    sample = list()
    ysample=list()
    n_sample = round(len(x) * ratio)
    while len(sample) < n_sample:
        index = randrange(len(x))
        sample.append(x[index])
        ysample.append(y[index])
    return sample,ysample

# Bootstrap Aggregation: Bagging
def bagging(X_train,y_train,X_test,sample_size=0.5):
    trees = list()
    for i in range(0,10):
        sample,ysample = subsample(X_train.values,y_train.values.reshape(-1,1), sample_size)
        reg = DecisionTreeRegressor(min_samples_split=3, max_depth=3)
        reg.fit(sample,ysample)
        trees.append(reg)
    predictions = [tree.predict(X_test.values) for tree in trees]
    predictions =pd.DataFrame(predictions).mean().values
    return(predictions)

preds=bagging(X_train,y_train,X_test, sample_size=0.5)
```

```python
# RMSE
print("RMSE: %.2f" % mean_squared_error(y_test, preds, squared=False))
# MAE
print("MAE: %.2f" % mean_absolute_error(y_test, preds))
```

Con las métricas calculadas, podemos concluir lo siguiente:

RMSE: El RMSE de 1683 es útil para comparar este modelo con otros, para el caso del árbol de decisión del punto anterior nos lleva a concluir que el modelo Bagging es ligeramente mejor.

MAE: El módelo de bagging arroja un MAE de 1253 dólares, esta precisión se considera útil ya que la media de los precios es de 14.538 dólares. Es decir, la desviación corresponde al 8.61% de la media.

### Punto 5 - Calibración de parámetros Random forest

En la celda 5, calibren los parámetros max_depth, max_features y n_estimators del modelo de Randon Forest para regresión, comenten sobre el desempeño del modelo y describan cómo cada parámetro afecta el desempeño del modelo.

```python
# Calibración n_estimators
# Creación de lista de valores para iterar sobre diferentes valores de n_estimators
estimator_range = range(10, 310, 10)

mean_squared_error(y_test, y_pred1, squared=False)
# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de n_estimators
accuracy_scores = []
mse_scores = []
# Uso de un 5-fold cross-validation para cada valor de n_estimators
for estimator in estimator_range:
    rrf1 = RandomForestRegressor(n_estimators=estimator, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(rrf1, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs la cantidad de n_estimators
plt.plot(estimator_range, accuracy_scores)
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
```

Se observa que a partir de 150 estimadores el accuracy no mejora significativamente, por lo que se decide tomar este valor como número de estimadores óptimo. Más estimadores no traeran mejora apreciable e incrementarán la demanda computacional.

```python
# Calibración max_features 
# Creación de lista de valores para iterar sobre diferentes valores de max_features
feature_cols = X.shape[1]
feature_range = range(1, feature_cols+1)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de max_features
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de max_features
for feature in feature_range:
    rrf2 = RandomForestRegressor(n_estimators=150, max_features=feature, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(rrf2, X, y, cv=5).mean())
    mse_scores.append(np.sqrt(np.abs(cross_val_score(rrf2, X, y, cv=5,scoring='neg_mean_squared_error').mean())))
```

```python
# Gráfica del desempeño del modelo vs la cantidad de max_features
plt.plot(feature_range, accuracy_scores)
plt.xlabel('max_features')
plt.ylabel('Accuracy')
```

Conforme se van añadiendo variables predictoras al modelo, el accuracy presenta leve mejoría. Por eso se decide tomar el número máximo de variables disponibles, en este caso max_features=9.

```python
# Calibración max_depth 
# Creación de lista de valores para iterar sobre diferentes valores de max_features
depth_range = range(1, 11)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de max_features
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de max_features
for depth in depth_range:
    rrf3 = RandomForestRegressor(n_estimators=150, max_features=9, max_depth=depth, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(rrf3, X, y, cv=5).mean())
    mse_scores.append(np.sqrt(np.abs(cross_val_score(rrf3, X, y, cv=5,scoring='neg_mean_squared_error').mean())))
```

```python
# Gráfica del desempeño del modelo vs la cantidad de max_depth
plt.plot(depth_range, accuracy_scores)
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
```

El parámetro de profundidad (max_depth) mejora sustancialmente el accuracy hasta cierto punto. A partir de apróximadamente el valor de 7 no se presenta mejoría notoria pues parece que se estabiliza. Por lo tanto, se decide tomar este valor como el número óptimo de profundidad.

```python
# Definición del modelo con los parámetros calibrados
crrf = RandomForestRegressor(n_estimators=150, max_features=9, max_depth=7, random_state=1, n_jobs=-1)
crrf.fit(X_train, y_train)
```

```python
#Impresión de desempeño del modelo
y_pred2 = crrf.predict(X_test)

# RMSE
crrf_RMSE = mean_squared_error(y_test, y_pred2, squared=False)
print("RMSE: %.3f" %crrf_RMSE )
# MAE
crrf_MAE = mean_absolute_error(y_test, y_pred2)
print("MAE: %.3f" %crrf_MAE )
```

Como se puede observar, los errores resultaron menores que los obtenidos sin calibrar el modelo. Está vez, el entrenamiento del modelo calibrado permite confirmar que el desempeño del modelo es mejor.

### Punto 7 - Calibración de parámetros XGBoost

En la celda 7 calibren los parámetros learning rate, gamma y colsample_bytree del modelo XGBoost para regresión, comenten sobre el desempeño del modelo y describan cómo cada parámetro afecta el desempeño del modelo.

```python
# Calibración n_estimators
# Creación de lista de valores para iterar sobre diferentes valores de n_estimators
estimator_range = range(10, 110, 5)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de n_estimators
accuracy_scores = []

# Uso de un 5-fold cross-validation para cada valor de n_estimators
for estimator in estimator_range:
    xgb1 = XGBRegressor(n_estimators=estimator, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb1, X, y, cv=5).mean())
```

Teniendo en cuenta el análisis anterior, se puede apreciar que el accuracy es cada vez mejor hasta más o menos 15 predictores y después disminuye, no mejora el poder predictivo del modelo. Por eso después se utilizan n_estimators=15.

```python
# Calibración subsample
# Creación de lista de valores para iterar sobre diferentes valores de n_estimators
subsample_range = np.arange(0.1, 1.1, 0.1)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de n_estimators
accuracy_scores = []

# Uso de un 5-fold cross-validation para cada valor de n_estimators
for subsample in subsample_range:
    xgb2 = XGBRegressor(n_estimators=15, subsample=subsample, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb2, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs subsample
plt.plot(subsample_range, accuracy_scores)
plt.xlabel('subsample')
plt.ylabel('Accuracy')
```

En el caso de subsample o proporción de submuestra de la instancia de entrenamiento, 0.9 resulta en el valor más alto para accuracy del modelo.

```python
# Calibración max_depth 
# Creación de lista de valores para iterar sobre diferentes valores de max_features
depth_range = range(1, 11)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de max_features
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de max_features
for depth in depth_range:
    xgb3 = XGBRegressor(n_estimators=15, subsample=0.9, max_depth=depth, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb3, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs max_depth
plt.plot(depth_range, accuracy_scores)
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
```

La profundidad máxima para el modelo del caso, es igual a 4 porque la exactitud del modelo presenta el mayor valor.

```python
# Calibración learning rate
# Creación de lista de valores para iterar sobre diferentes valores de learning_rate
learning_rate_range = np.arange(0.1, 1.1, 0.1)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de learning_rate
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de learning_rate
for learning_rate in learning_rate_range:
    xgb3 = XGBRegressor(n_estimators=15, subsample=0.9, max_depth=4, learning_rate=learning_rate, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb3, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs learning_rate
plt.plot(learning_rate_range, accuracy_scores)
plt.xlabel('learning_rate')
plt.ylabel('Accuracy')
```

El learning_rate óptimo es igual a 0.3 como el valor default para la librería de acuerdo al accuracy que resulta para el desempeño del modelo.

```python
# Calibración gamma
# Creación de lista de valores para iterar sobre diferentes valores de gamma
gamma_range = range(1, 100, 5)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de gamma
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de gamma
for gamma in gamma_range:
    xgb3 = XGBRegressor(n_estimators=15, subsample=0.9, max_depth=4, learning_rate=0.3, gamma=gamma, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb3, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs gamma
plt.plot(gamma_range, accuracy_scores)
plt.xlabel('gamma')
plt.ylabel('Accuracy')
```

Por otra parte el valor gamma o reducción de pérdida mínima necesaria para realizar una partición adicional en un nodo de hoja del árbol, es estable y por eso se deja el vlaor por default que es igual a 0.

```python
# Calibración colsample_bytree
# Creación de lista de valores para iterar sobre diferentes valores de colsample_bytree
colsample_bytree_range = np.arange(0.1, 1.1, 0.1)

# Definición de lista para almacenar la exactitud (accuracy) promedio para cada valor de colsample_bytree
accuracy_scores = []

# Uso de un 10-fold cross-validation para cada valor de colsample_bytree
for colsample_bytree in colsample_bytree_range:
    xgb3 = XGBRegressor(n_estimators=15, subsample=0.9, max_depth=4, learning_rate=0.3, gamma=0,
                        colsample_bytree=colsample_bytree, random_state=1, n_jobs=-1)
    accuracy_scores.append(cross_val_score(xgb3, X, y, cv=5).mean())
```

```python
# Gráfica del desempeño del modelo vs colsample_bytree
plt.plot(colsample_bytree_range, accuracy_scores)
plt.xlabel('colsample_bytree')
plt.ylabel('Accuracy')
```

Finalmente para el parámetro colsample_bytree que se refiere a la relación de submuestra de columnas al construir cada árbol, se selecciona el valor 0.6 porque después el accuracy se hace estable. Después, se corre el modelo con los parámetros calibrados y podemos observar el desempeño del modelo.

```python
# Definición del modelo con los parámetros calibrados
cxgb = XGBRegressor(n_estimators=15, subsample=0.9, max_depth=4, learning_rate=0.3, gamma=0, colsample_bytree=0.6,
                    random_state=1, n_jobs=-1)
cxgb.fit(X_train, y_train)
```

```python
#Impresión de desempeño del modelo
y_pred4 = cxgb.predict(X_test)

# RMSE
cxgb_RMSE = mean_squared_error(y_test, y_pred4, squared=False)
print("RMSE: %.3f" %cxgb_RMSE )
# MAE
cxgb_MAE = mean_absolute_error(y_test, y_pred4)
print("MAE: %.3f" %cxgb_MAE )
```

El modelo XGBoost con los parámetros calibrados presenta RMSE y MSE similares al XGBoost y random forest calibrado, es ligeramente mejor que el primero sin embargo nuevamente no es mejor que el segundo.
